# 2. 知性を表現する手法ーゲームAIの基本概念ー



## デジタルゲームにおける人工知能

一般社会における人工知能が果たすべき役割と、ゲームにおける人工知能が果たすべき役割は異なる。

ゲームにおける人工知能が果たすべき役割とは「ゲームをおもしろくすること」
* 特にこのようなAIを`エンターテイメントAI`と呼ぶことも多い。

### ゲームキャラクターに必要な要素
* 演技
* リアリティー

2Dのゲームまでは、マップを俯瞰的に見た状況判断でプログラムが書かれるため、いわば操り人形のように動いていた。
* そこには、世界（環境）の認識や、思考、行動の出力といった、自律型AIがもつ流れは一切介在しなかった。
3Dのゲーム開発が盛んになった今、マップを俯瞰的にみた処理でプログラムを書くのは（不可能ではないが）ちょいと困難であるため、キャラクターAI自身が、世界（環境）を認識に、状況判断（思考）し、行動する必要性があった。

### より人間らしく-自律型AI-

ゲームにおける自律型キャラクターAIに求められる要素として、これまで述べたように「環境を認識して自律的に行動する能力」と「ゲームに則した演技をする能力」がある。

しかし、これら2つの能力は、知能を再現するという人工知能の領域において矛盾したものであり、ゲームやエンターテイメントAIに求められる大きな特徴でもある。

## 現実感を追及する-不気味の谷を超えて-

2000年以降のハードウェアも含めたゲーム開発技術の向上により、人間のプレイヤーがゲーム世界を、現実（リアリティー）な世界として錯覚、没入するような仕組み、技術が多く投入された。

そんななかで、人間の姿を模したキャラクターAIの精巧さが向上される中で、よりリアリティーが求められるように。
* ある意味では「不気味の谷」を超えたともいえる。
* 不気味の谷とは、より人間に近いキャラクターが、人間らしい振る舞いを見せれなかったときに起こる、不気味さ、嫌悪感のこと。

人間はさも当然のように障害物をよけながら運動することができるが、AIにはそれは難しい。かといって、そのような課題を直接的にクリアするようなコードを書くとなると、どのような状況でも、していされた動作をしてしまうので、人間のような柔軟性に欠ける。

人間にとっては当たり前にできることの再現上に、さらに高度な思考機能を実装することでリアリティのある知性に近づけることができる。

> ゲームAIを作るとは、そういった人間の無意識的に行える行動を、明示的にプログラミングしてあげることに他ならない。


* 知性の再現からくるリアリティー、ゲームに則した演技をすることで得られるリアリティーこの2つはどちらがどれほど大切なのか。
  * 答えは、両輪であるといえる。
  
日本と西洋ではこの部分でも大きな差がある。

海外ではオープンワールドなゲームが流行った理由として、現実の世界をデジタル内に再現する（あるいはされる）ことに対して異様な喜びと重視がある、一方日本では、ゲームの世界は現実とは異なるべつの空間（仮想世界）という意味合いが強く、キャラクターそのものの知性部分よりも、演技の部分が重要視される傾向にある。

ただし、近年のグラフィックスの精巧さから、日本のキャラクターにも、演技以外の知性のクオリティが求められるようになってきており、両者のリアリティーの重要さは均衡している。

## 3つのゲームAIでユーザー体験をつくる

近年のゲームにおけるAIは、以下の3つの機能に大別される。

* メタAI
  * ゲーム全体を監視し、状況に応じてゲーム内の世界のパラメータを調整する役割を担うAI
* ナビゲーションAI
  * 後述するキャラクターAIに、ゲーム内の地形情報,位置情報などなどを提供するAI
* キャラクターAI
  * NPCや敵役など、用途は多岐にわたる。ゲーム内世界を自律的に行動し、ときにプレイヤーの仲間になって協調し行動する役割を持つ。


## 生きているということ-「らしさ」を作る-

ディスプレイの中にいるキャラクターが、生き生きと会話をしてくれる、そんな世界を実現するのがゲーム開発者の目標でもある。

RPGなどで多いが、キャラクター同士の会話の場面で、呼吸による振動、集団で会話をするときに無意識的に形成される社会的空間を再現することで、より人間らしい振る舞いを与える事ができる。

特に、現実世界のわれわれ人間は無意識的に作り出す空間・範囲のことを「社会的空間」と呼び、これには`O-Space`,`R-Space`,`P-Space`という3つの空間がある。
* O-Space
  * その人のいわばなわばり的空間、パーソナルスペースとも呼ばれる。
* P-Space
  * 複数人で会話をしているときに形成される空間、グループごとに形成される。
* R-Space
  * P-Spaceの外側、グループと外界の境界線を形成する空間。

このR-Spaceに他人が入り込むと、人間は本能的にその人物に注意を払う事が多い。視線を向けたりなど。
このような動きをキャラクターに付けることで、より人間らしい振る舞いを与えることができる。

キャラクター同士の社会的関係を表現するにあたり、上のような社会的スペースの考え方は重要である。

ハイエナや犬などの動物は、自身のもつ範囲（なわばり）に、未知の生物（他の仲間でも良い）が侵入したときに彼らは怒りをあらわにする。
このように、人間をはじめ、多くの動物は自身のなわばりに他人が侵入したときに、不快感・怒りの感情をあらわにする、という理論を`アージ理論`と呼ぶ。

このアージ理論をはじめ、「場所・距離」というのは社会的な意味を含んでいる。

### 役割と役割に応じた行動

キャラクターの役割を明確に定めることは、キャラクターAIを開発するさいには有効
* そのAIが解くべきフレーム（課題、役割、仕事）が限定されるので、フレーム問題を回避できる。

一般的なゲームの場合、キャラクターAIの役割を動的に変化させることは、ゲームのシナリオに影響するためあまり見られないが、変更したいばあいはメタAIから変更するのが一般的。

## メタ思考とメタ知識-思考を組み立てる-

メタ知識とは、「ある事について知っている」ということを知っている、つまりその知識について知っている事を指す
* 例えば、リンゴの色を知っている、雨が降る前の雲の形を知っている、などなど
* 要するに、具体的な答え（値）の一段上、その答えを知っているという事を知っていることを指す。

メタ知識があれば、知らない答えをどう集めれば良いかを知ることができ、キャラクターの行動分岐に大いに影響する。

## ブラックボードで情報を把握する
黒板は、情報を自由に書いてそして消すことができる。
AIにおいて、この黒板の概念は重要

黒板を階層化して情報を把握することを「階層化黒板」と呼ぶ。

階層化黒板は、いくつかのレイヤごとに情報を整理し、上の層に行くほど、下の層の情報を抽象化した情報に整理するというもの。

この各層ごとに配置されている黒板の情報を上位層の黒板へ情報を整理するブロックのことを「ナレッジベース」と呼ぶ。

ゲームでは、このように観測情報を抽象化して、全体の状況把握に役立てることができる。

このように階層化黒板は、観測された情報を収集し、まだ知られていないより高度な情報を得ることができる。
しかし、このような情報処理は、ある一部分に特化したものになるため、状況の変化などに融通が効きにくい。

## ブラックボードアーキテクチャで世界を認識する

観測した情報を、知識として抽象化し整理する。このブラックボードアーキテクチャの考えは、現在のAIエージェントの学習方法に置ける基礎を形成している。
