# 2. 知性を表現する手法ーゲームAIの基本概念ー



## デジタルゲームにおける人工知能

一般社会における人工知能が果たすべき役割と、ゲームにおける人工知能が果たすべき役割は異なる。

ゲームにおける人工知能が果たすべき役割とは「ゲームをおもしろくすること」
* 特にこのようなAIを`エンターテイメントAI`と呼ぶことも多い。

### ゲームキャラクターに必要な要素
* 演技
* リアリティー

2Dのゲームまでは、マップを俯瞰的に見た状況判断でプログラムが書かれるため、いわば操り人形のように動いていた。
* そこには、世界（環境）の認識や、思考、行動の出力といった、自律型AIがもつ流れは一切介在しなかった。
3Dのゲーム開発が盛んになった今、マップを俯瞰的にみた処理でプログラムを書くのは（不可能ではないが）ちょいと困難であるため、キャラクターAI自身が、世界（環境）を認識に、状況判断（思考）し、行動する必要性があった。

### より人間らしく-自律型AI-

ゲームにおける自律型キャラクターAIに求められる要素として、これまで述べたように「環境を認識して自律的に行動する能力」と「ゲームに則した演技をする能力」がある。

しかし、これら2つの能力は、知能を再現するという人工知能の領域において矛盾したものであり、ゲームやエンターテイメントAIに求められる大きな特徴でもある。

## 現実感を追及する-不気味の谷を超えて-

2000年以降のハードウェアも含めたゲーム開発技術の向上により、人間のプレイヤーがゲーム世界を、現実（リアリティー）な世界として錯覚、没入するような仕組み、技術が多く投入された。

そんななかで、人間の姿を模したキャラクターAIの精巧さが向上される中で、よりリアリティーが求められるように。
* ある意味では「不気味の谷」を超えたともいえる。
* 不気味の谷とは、より人間に近いキャラクターが、人間らしい振る舞いを見せれなかったときに起こる、不気味さ、嫌悪感のこと。

人間はさも当然のように障害物をよけながら運動することができるが、AIにはそれは難しい。かといって、そのような課題を直接的にクリアするようなコードを書くとなると、どのような状況でも、していされた動作をしてしまうので、人間のような柔軟性に欠ける。

人間にとっては当たり前にできることの再現上に、さらに高度な思考機能を実装することでリアリティのある知性に近づけることができる。

> ゲームAIを作るとは、そういった人間の無意識的に行える行動を、明示的にプログラミングしてあげることに他ならない。


* 知性の再現からくるリアリティー、ゲームに則した演技をすることで得られるリアリティーこの2つはどちらがどれほど大切なのか。
  * 答えは、両輪であるといえる。
  
日本と西洋ではこの部分でも大きな差がある。

海外ではオープンワールドなゲームが流行った理由として、現実の世界をデジタル内に再現する（あるいはされる）ことに対して異様な喜びと重視がある、一方日本では、ゲームの世界は現実とは異なるべつの空間（仮想世界）という意味合いが強く、キャラクターそのものの知性部分よりも、演技の部分が重要視される傾向にある。

ただし、近年のグラフィックスの精巧さから、日本のキャラクターにも、演技以外の知性のクオリティが求められるようになってきており、両者のリアリティーの重要さは均衡している。

## 3つのゲームAIでユーザー体験をつくる

近年のゲームにおけるAIは、以下の3つの機能に大別される。

* メタAI
  * ゲーム全体を監視し、状況に応じてゲーム内の世界のパラメータを調整する役割を担うAI
* ナビゲーションAI
  * 後述するキャラクターAIに、ゲーム内の地形情報,位置情報などなどを提供するAI
* キャラクターAI
  * NPCや敵役など、用途は多岐にわたる。ゲーム内世界を自律的に行動し、ときにプレイヤーの仲間になって協調し行動する役割を持つ。
