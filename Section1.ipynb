# 1章 知能って何だろう。-自然知能と人工知能-

## 知能と宇宙

> 「知能」とは何かについて、我々はその答えをまだもってはいない。

### 知能を定義する

* 自分の内部を外部から守る、防衛すること。自身の欲求を環境に反映させること。
  * 例えば我々人間には、食欲・睡眠欲・性欲といった欲求があり、日々の生活の中でそれらの欲求を満たす
  * >　知能は、身体と環境の境界上にあり、外部から我々の内部を守るもの。
* 外部と内部の境界でその関係性を調整するもの。
  * > 身体の状況を察知し、その状況を緩和するために行動を起こす。原始的な知能の役割
* 知能は身体と環境の関係をとりもつ

* 動物のもつ、「なわばり」的なものが感情の源、源泉であるという考えもある：アージ理論と呼ぶ。

> 身体と環境は特有のつながりを持ち、知能はその身体に内在するものである。

* 知能は石や木と同じく自然的に発生したもの。ゆえに形、形式は様々だが共通した性質というものがある。
  > 知能とは生き物に共通した性質

* このことを「知能の原理」と呼び、理学はそれを追及する学問
  
## エンジニアリングとしての人工知能

* 知能の原理が、人工知能を動かす材料であり、自然界に存在する様々な知能をコードに落とし込んで、人工的にそれを再現することである。

* 知能とは「宇宙」の対義語である
  
* 人工知能の黎明期において、知能とは「言葉を操り推論する能力をもつもの」と定義され、かなり高い能力を求めていた

* 西洋の人々は、知能＝言葉を操るという考えが露骨に強く、ChatGPTをはじめ、WatsonやGoogleの検索エンジンなど、「言葉、言語」に重きを置いた人工知能が開発されてきた。
* 対して日本、とりわけ東洋ではそこまで言葉＝知能という側面は重要視されなかった。
  * 日本では「八百万の神」という言葉があるように、知能の本質は言葉にあるものではないという考え方が強かった。
  
## AIの歴史とゲームAIの歴史

* 人工知能の学問が始まったのは1950年代後半であり、数学や物理学といった学問よりもはるかに新しい学問である。
* そしてコンピュータという「人間の知的作業を補助する機械」が登場し、社会に浸透してきたのが1980, 90年代である。

* 人工知能には人間の知能を再現するという明確な方向性がありながら、その方角は多岐に及ぶことから（人間の知能は色々できる）可能性が無限大にあることは容易に想像できる。

### ファミコン登場以前のゲームAI（1980年以前）

* 当時はまだ秋葉原で基盤やROMなどが露店で売られていた時代。
* またプログラミングもそんなに普及していなかった（アセンブリ言語が主役の時代）ため、コーディングのむずかしさ、計算リソースの制約などにより、ゲームAIは非常にシンプルなもので、単に「敵役」「標的」という意味合いしかもてなかった。
  
### ファミコン登場以降のゲームAI（1980年以降）

* RPGやアクションゲームの登場によって、単に敵としてではなく、自身をサポートする仲間、味方としても登場するようになった。
* ただこれも、大々的に知能を再現したものではなく、あくまでゲーム処理の中に、ゲームを盛り上げるスパイスといった意味合い程度の処理がなされていたに過ぎない。

### 3Dゲームの時代

* 近年のＦＰＳゲームに代表されるような、２Ｄとは異なる空間でのゲーム開発なので、ゲームAIの開発においても、大きな変化が起きている。
  * ２Ｄでは、プレイヤーや敵の位置を俯瞰的に見ることができるので、ＡＩが自律的に動くというよりは、プログラムで動作シナリオを描いて、動かすようなものであった。
  * ３Ｄでは、プレイヤーキャラクターの視点のみからユーザーは見ることができない。そのため、AIの処理方法を従来通り、シナリオを描いてプログラムで再現するのが困難になった。
* ナビメッシュなどの技術の登場により、AIはゲーム空間内を自律的に動き回れるようになった。

ゲームにおける人工知能は、自身が存在する環境の認識、評価、行動、行動が環境に与えた影響を観察するというサイクルを繰り返す事で知能としての側面をもつ。
